```python
import pandas as pd
import matplotlib.pyplot as plt


datos=pd.read_csv("products.csv")
df=datos[["productID","categoryID"]]

print (df.sample(5))

valor_por_producto = df.groupby("productID")["categoryID"].mean()
valor_por_producto.head(10).plot.barh()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


datos=pd.read_csv("products.csv")
df=datos[["unitPrice","categoryID"]]

print (df.sample(5))
df.unitPrice.value_counts().head(10).plot.pie()
plt.show()
```
